```python
import requests
from datetime import datetime, timedelta
import random
import pickle
import json
import os
from time import sleep
import pandas as pd

def fetch_data():

    endpoint = "https://rickandmortyapi.com/api/character"
    endpoint2 = "https://rickandmortyapi.com/api/location"
    endpoint3 = "https://rickandmortyapi.com/api/episode"
    
    #Recogemos las respuestas de los 3 ENDPOINTS
    response = requests.get(endpoint)
    response2 = requests.get(endpoint2)
    response3 = requests.get(endpoint3)
    
    datos1 = response.json()
    datos2 = response2.json()
    datos3 =response3.json()
    data = (datos1,datos2,datos3)

    return data

def trasnform_data(total_data):
    
    data_characters, data_locations, data_episodes = total_data
    
    df_characters = pd.DataFrame(data_characters["results"])
    
    df_characters = fix_col_dict(df_characters, "origin")
    df_characters = fix_col_dict(df_characters, "location")
    #Borramos columnas que no nos interesan
    df_characters = df_characters.drop(["image","episode","url", "created"],axis = 1)
    
    df_locations = pd.DataFrame(data_locations["results"])
    
    df_episodes = pd.DataFrame(data_episodes["results"])
    
    return df_characters, df_locations, df_episodes


def fix_col_dict(data_frame: pd.DataFrame, nombre_col):
    
    datos =[]
    
    for fila in data_frame[nombre_col]:
        datos.append(fila["name"])
        
    data_frame[f"{nombre_col}_nueva"] = datos
    
    # Borramos la columna y la devolvemos
    return data_frame.drop(nombre_col, axis=1)
    
        

#EJECUCION 
data = fetch_data()
df_characters, df_locations, df_episodes = trasnform_data(data)

df_characters

```
